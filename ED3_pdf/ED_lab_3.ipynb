```python
number_of_points=1000
x_point = []
y_point = []
x_1=[]
y_1=[]
```

```python
a=0.22
b=0.78
```

```python
for i in range(number_of_points):
  x = np.random.normal(0.0,0.5)
  y = a*x+b+np.random.normal(0.0,0.1)
  x_point.append(x)
  y_point.append(y)
  x_1.append(1)
  y_1.append(1)
```

```python
plt.scatter(x_point,y_point,c='b')
plt.show()
```

```python


df = pd.DataFrame({"x" : x_point, "y" : y_point})
df.to_csv("punkty2D.csv", index=False)
```

Zadanie 6

```python
x_s=np.array(x_point)
y_s=np.array(y_point)


x2=np.sum(x_s*x_s)
x1=np.sum(x_s)
x1sum=np.sum(x_1)
y1sum=np.sum(y_1)
xy=np.sum(x_s*y_s)
y1=np.sum(y_s)
```

```python
print(x1)
```

```python
M = np.array([[x2,x1],[x1,x1sum]])
print(M)
```

```python
M_1=np.linalg.inv(M)
print(M_1)
```

```python
N=np.array([[xy],[y1]])
print(N)
```

```python
a,b = np.matmul(M_1,N)
print(a,b)

```

```python
plt.scatter(x_s,y_s)
plt.plot(x_s,a*x_s+b,"black")
plt.show()
```
