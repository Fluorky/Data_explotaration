Zadanie 3

```python
import matplotlib.pyplot as plt
```

```python
rm=pd.DataFrame(data,columns=['rm'])
medv=pd.DataFrame(data,columns=['medv'])
```

```python
plt.scatter(rm,medv)
plt.show()
```

```python
number_of_points=506
x_point = []
y_point = []
x_1=[]
y_1=[]
```

```python
for i in range(number_of_points):
  x_1.append(1)
  y_1.append(1)
```

```python
x_s=np.array(rm)
y_s=np.array(medv)


x2=np.sum(x_s*x_s)
x1=np.sum(x_s)
x1sum=np.sum(x_1)
y1sum=np.sum(y_1)
xy=np.sum(x_s*y_s)
y1=np.sum(y_s)



print(x2,x1,x1sum,y1sum,xy,y1)
```

```python
M = np.array([[x2,x1],[x1,x1sum]])
```

```python
M_1=np.linalg.inv(M)
```

```python
N=np.array([[xy],[y1]])
print(N)
```

```python
a,b = np.matmul(M_1,N)
print(a,b)
```

```python
plt.scatter(x_s,y_s)
plt.plot(x_s,a*x_s+b,"black")
plt.show()
```
